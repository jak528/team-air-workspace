```python
Public_Parking_JFK=JFK[['Public_Parking_Counts','Year','Total_Passengers']]
print(Public_Parking_JFK)
```

```python
Parking_JFK1 = [2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015]
```

```python
JFK_data_year=[]     # JFK airport data for specific year
for i in Parking_JFK1:
    JFK_data_year.append(Public_Parking_JFK.loc[Public_Parking_JFK['Year']==i])
print(JFK_data_year)
```

```python
print(JFK_data_year[0])
```

```python
parking_sum_year=[]
passengers_sum_year=[]
parking_ratio=[]
for i in JFK_data_year:
    temp =i
    parking_sum_year.append(temp.Public_Parking_Counts.sum())
    passengers_sum_year.append(temp.Total_Passengers.sum())
    parking_ratio.append(temp.Public_Parking_Counts.sum()/temp.Total_Passengers.sum())
```

```python
print(parking_sum_year)
print(passengers_sum_year)
print(parking_ratio)
```

```python

plt.plot(Parking_JFK1,parking_ratio)# ration of Public Parking in total passengers for JFK
plt.xlabel('Year')
plt.ylabel('Public Parking Counts for JFK Airport')
plt.show()
```

```python
Taxi_JFK=JFK[['Taxi_Dispatched','Year','Total_Passengers']]
Parking_JFK2 = [2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014]
JFK_data_year2=[]     # JFK airport data for specific year
for i in Parking_JFK:
    JFK_data_year2.append(Taxi_JFK.loc[Taxi_JFK['Year']==i])
```

```python
Taxi_sum_year=[]
passengers_sum_year=[]
Taxi_ratio=[]
for i in JFK_data_year2:
    temp =i
    Taxi_sum_year.append(temp.Taxi_Dispatched.sum())
    passengers_sum_year.append(temp.Total_Passengers.sum())
    Taxi_ratio.append(temp.Taxi_Dispatched.sum()/temp.Total_Passengers.sum())
```

```python
plt.plot(Parking_JFK2,Taxi_ratio)     # ration of Taxi Dispatched for JFK
plt.xlabel('Year')
plt.ylabel('Taxi Dispatched for JFK Airport')
plt.show()
```
